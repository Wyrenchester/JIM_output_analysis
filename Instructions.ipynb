# JIM Output Analysis

#### This is an example of how to run the python code. Files (named 'Green' and 'Red') can be found in the example folder if you'd like to follow along.

##### The example files are from a lysis experiment performed on HIV-1 virions on coverglass. It has 50 total particles. The channels are:
&nbsp; &nbsp; Red: HIV-1 capsid marker 

&nbsp; &nbsp; Green: Fluid phase marker

### 1. Initialization

Install libraries as needed. To install dependencies with pip, run: *pip install numpy pandas matplotlib scipy hmmlearn xlsxwriter*

Hmmlearn ddepends on other libraries (numpy, scipy, and scikit-learn). These should be installed automatically if you use pip or conda.

### 2. Run code
To run the code, press the 'Run File' button. (The example below is in Spyder.)

### 2. GUI input

A GUI should appear (it may appear behind the Python window). The descriptions for each field are as follows:

| Field                               | Description                                                                                                                                                        |
| ----------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **Red Channel File**                | Select the Excel or CSV file containing Red channel intensity data exported from JIM. Use the file picker or enter the path manually.                              |
| **Green Channel File**              | Select the Excel or CSV file containing Green channel intensity data exported from JIM.                                                                            |
| **HMM Fit Channel**                 | Choose which channel (Red or Green) will be analyzed using Hidden Markov Models (HMM). The chosen channel is used to detect stage changes like uncoating or lysis. |
| **Alignment Channel** *(optional)*  | If alignment analysis is enabled, choose which channel will be aligned to the second drop in the HMM-fit channel.                                                  |
| **Save Plot**                       | Select whether to save plots of particle traces and HMM fits. If yes, a folder with subfolders for particle categories will be created.                            |
| **Seconds per frame**               | Enter the time interval between frames (in seconds). This is used to convert frames to minutes in plots and Excel outputs.                                         |
| **Frames before drop** *(optional)* | Number of frames prior to the second drop to include when aligning traces.                                                                                         |
| **New filename**                    | Specify a base name for output files. Defaults to the Red or Green file name if left blank.                                                                        |

After specifing each of the following, hit the **'Update Parameters'** button to apply settings to the analysis.

The message box should now state: *Parameters updated: HMM=red, SavePlot=y, Align=y, AlignChan=green, FramesBefore=5, Output=Red*

Then, hit the **'Process'** button to process the data. A window will pop up. Specify a folder to save results in, and hit **'Select Folder'**. *You should already be in the data folder, so no selection should be required.*

If you'd like to confirm that the code is running correctly, the consol should output *'Processing particle x/y...'*, where *x* is the particle being processed and *y* is the total number of particles.

After processing, the consol will output:

&nbsp; &nbsp; *HMM complete. Writing Excel outputs...*

&nbsp; &nbsp; *Analysis complete. Files saved in: /file_path*

Hit **'Finish'** to end the program.

&nbsp; &nbsp; *Make sure that the console says: Analysis complete. Files saved in: file_path **before** exiting!*

### 3. Outcome interpretation

Output excels are in the 'Red_plots' > 'results' folder. 

Graphs are in folders based on their HMM fit:

&nbsp; &nbsp; Lysis: single drop

&nbsp; &nbsp; Nonlytic: no drops

&nbsp; &nbsp; Uncoating: two drops
